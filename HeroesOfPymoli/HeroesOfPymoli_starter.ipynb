## Gender Demographics

* Percentage and Count of Male Players
purchase_data["Gender" = "Male"].count()

* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

CntMale = purchase_data["Gender"].value_counts()['Male']
CntFemale = purchase_data["Gender"].value_counts()['Female']
CntOther = purchase_data["Gender"].value_counts()['Other / Non-Disclosed']
CntTot = CntMale + CntFemale + CntOther
print(
    f" Total: {CntTot}\n Male: {CntMale}\n Female: {CntFemale}\n non_specfic: {CntOther}")
print(" ")
PctMale = (CntMale / CntTot) * 100
PctFemale = (CntFemale / CntTot) * 100
PctOther = (CntOther / CntTot) * 100
print(
    f" % Male: {PctMale}\n % Female: {PctFemale}\n % non_specifc: {PctOther}")
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
OnlyMale = purchase_data.loc[purchase_data["Gender"] == "Male", :] 
SumMale = OnlyMale["Price"].sum()
AveMale = OnlyMale["Price"].mean()
CntMaleSN = len(OnlyMale["SN"].unique())
AvePerMale = SumMale/CntMaleSN
print(
    f" Male Total: {CntMale}\n Male AvePrice: {AveMale}\n Male TotPrice: {SumMale}\n Male AvePerPerson: {AvePerMale}")
print(" ")
OnlyFemale = purchase_data.loc[purchase_data["Gender"] == "Female", :] 
SumFemale = OnlyFemale["Price"].sum()
AveFemale = OnlyFemale["Price"].mean()
CntFemaleSN = len(OnlyFemale["SN"].unique())
AvePerFemale = SumFemale/CntFemaleSN
print(
    f" Female Total: {CntFemale}\n Female AvePrice: {AveFemale}\n Female TotPrice: {SumFemale}\n Female AvePerPerson: {AvePerFemale}")
print(" ")
OnlyOth = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :] 
SumOth = OnlyOth["Price"].sum()
AveOth = OnlyOth["Price"].mean()
CntOthSN = len(OnlyOth["SN"].unique())
AvePerOth = SumOth/CntOthSN
print(
    f" Other Total: {CntOther}\n Other AvePrice: {AveOth}\n Other TotPrice: {SumOth}\n Other AvePerPerson: {AvePerOth}")
print(" ")
# Create a new frame consolodating above calculations

# Creating a Pandas DataFrame by passing in a LIST OF DICTIONARIES
# Each value in the list is a dictionary
# Imagine that each dictionary represents a row of data in our eventual dapurchase count, avg. purchase price, avg. purchase total per persontaframe
# Each dictionary should have the same keys, since these keys dictate the column headers of our dataframe
#                        purchase count, avg. purchase price, avg. purchase total per person
my_list = [{"Gender": "Male", "Purchase Count": CntMale, "Average Price Paid": AveMale, "Average Purchase PerPerson": AvePerMale},
           {"Gender": "Female", "Purchase Count": CntFemale, "Average Price Paid": AveFemale, "Average Purchase PerPerson": AvePerFemale},
           {"Gender": "Other / Non-Disclosed", "Purchase Count": CntOther, "Average Price Paid": AveOth, "Average Purchase PerPerson": AvePerOth}]
ConsolPdDf = pd.DataFrame(my_list)
ConsolPdDf
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins and bin labels for the age column

Age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
Age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Bin the Age column
# cut() returns a Pandas Series containing each of the binned column's values translated into their corresponding bins

#pd.cut(purchase_data["Age"], Age_bins, labels=Age_labels)

# We can append our bins to Aged_df
Aged_df = purchase_data
Aged_df["Level"] = pd.cut(purchase_data["Age"], Age_bins, labels=Age_labels)
#Aged_df
grouped_Aged_df = Aged_df.groupby("Level")
TotByLevel_df = grouped_Aged_df[["Item ID"]].count()
TotByRen_df = TotByLevel_df.rename(columns={"Item ID":"Total Count"})

TotByLvl_df = TotByRen_df
TotByLvl_df["Percentage Of Players"] = (TotByRen_df["Total Count"]/CntTot)

TotByLvl_df["Percentage Of Players"] = (TotByRen_df["Total Count"]/CntTot)
TotByLvl_df.style.format({'Total Count': "{:.0f}",'Percentage Of Players': "{:.2%}"})

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#    ----------------------from above
# Aged_df
GrpAged_df = Aged_df.groupby("Level")

GrpCntByLvl = GrpAged_df
GrpCntByLvl["Purchase Count"] = GrpAged_df[["Item ID"]].count()
GrpCntByLvla = pd.DataFrame(GrpCntByLvl.drop(['Purchase ID', 'SN', 'Age', 'Gender', 'Item ID', 'Item Name', 'Price'], axis=1))

GrpAveByLvl
GrpAveByLvl["Average Purchase Price"] = GrpAged_df[["Price"]].mean()
GrpAveByLvla = pd.DataFrame(GrpAveByLvl.drop(['Price'], axis=1))

GrpSumByLvl
GrpSumByLvl["Total Purchase Value"] = GrpAged_df[["Price"]].sum()
GrpSumByLvla = pd.DataFrame(GrpSumByLvl.drop(['Price'], axis=1))

#GrpPerPersByLvl["Ave Total Purchase PerPerson"] = len(GrpAged_df('SN').agg(['nunique']))
#GrpPerPersByLvl 

# Merge the four created data frames on the (age)Level
# first the count and ave
merged_df = pd.merge(GrpCntByLvla, GrpAveByLvla, on="Level")

# then the sum
Merged2_df = pd.merge(merged_df, GrpSumByLvla, on="Level")

# finally the PerPerson   '${:.2f}'.format
PurchSummary_df = Merged2_df
#
PurchSummary_df.style.format({'Purchase Count': "{:.0f}",
                              'Average Purchase Price': "${:.2f}",
                              'Total Purchase Value': "${:.2f}"
                             })
```
